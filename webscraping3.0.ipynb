```python
#defining the para function 
def para_fxn(x):
    para_list = []
    soup = BeautifulSoup(x , 'html.parser')
    para = soup.find_all('p')
    for i in para:
        para_list.append(i.text)
    return para_list



#defining the h1 heading funtion 
def h1_fxn(x):
    head1_list = []
    soup = BeautifulSoup(x , 'html.parser')
    head = soup.find_all('h1')
    for i in head:
        head1_list.append(i.text)
    return head1_list



#defining the h2 heading funtion 
def h2_fxn(x):
    head2_list = []
    soup = BeautifulSoup(x , 'html.parser')
    head = soup.find_all('h2')
    for i in head:
        head2_list.append(i.text)
    return head2_list


#defining the span funtion 
def span_fxn(x):
    span_list = []
    soup = BeautifulSoup(x , 'html.parser')
    head = soup.find_all('span')
    for i in head:
        span_list.append(i.text)
    return span_list


#defining the h5 heading funtion 
def h5_fxn(x):
    head5_list = []
    soup = BeautifulSoup(x , 'html.parser')
    head = soup.find_all('h5')
    for i in head:
        head5_list.append(i.text)
    return head5_list


#defining div function 
def div_fxn(x):
    div_list = []
    soup = BeautifulSoup(x , 'html.parser')
    head = soup.find_all('div')
    for i in head:
        div_list.append(i.text)
    return div_list



#defining the h3 heading funtion 
def h3_fxn(x):
    head3_list = []
    soup = BeautifulSoup(x , 'html.parser')
    head = soup.find_all('h3')
    for i in head:
        head3_list.append(i.text)
    return head3_list
```
